We will need the probability density function for quite a few of the results, so lets write one.

```python
def norm(x, mu, sigma):
    ''' Return a normal distribution curve given mean and standard deviation.
    
        Args:
            x (list):        list of x values to graph curve over.
            mu (float):      mean 
            sigma(float):    standard deviation
            
        Returns:
            list:            list of y values for normal distribution
    '''
    
    coeff = 1.000 / np.sqrt(2.000 * np.pi * sigma * sigma)
    exp   = np.exp(-1.000 * (np.power(x - mu, 2) / (2.000 * sigma * sigma)))
    
    return coeff * exp 
```

```python
from scipy.fftpack import fft

# Calculate period.
T = t[1] - t[0]

# Count samples.
N = len(t)

# Find frequency bins according to Nyquist theorem.
x = np.linspace(0.0, 1.0 / (2.0 * T), int(N/2))

# Comput fast fourier transform (FFT)
yr = fft(v)
y = (2.0 / N) * np.abs(yr[0:np.int(N / 2.0)])

# Plot.
plt.plot(x, [i * 1000 for i in y], color='k')

# Set graph parameters.
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (mV)')
plt.grid(linestyle=':')

# Mark suspected 60 Hz noise (and its higher harmonics).
plt.xticks([i * 60 for i in range(20)])
plt.xlim(0, 800)
plt.ylim(0, 15)

# Show the graph.
plt.savefig('FFT', dpi=300)
plt.show()
```

```python
def histogram_stack(files, labels=None, start=-8, stop=8, step=0.25, save=None):
    ''' Plot all files as a stack of histograms with labels. '''
    
    # Make sure there are enough labels for each file.
    assert len(files) == len(labels), "Each file must have exactly one lable."
    
    N = len(files)
    
    # Create subplots.
    f, axes = plt.subplots(N, sharex=True, figsize=(5, int(N * 6 / 3)), dpi=120)
    f.subplots_adjust(hspace=0.125)
    
    # Set bin width to 25 mV (works better than actual algorithm...)
    bins = np.arange(start, stop, step)
    
    # Plot each plot in the stack.
    for axis, file, label in zip(axes, files, labels):
        # Import the file.
        df = pd.read_csv(file, header=None)
        df = np.asarray(df)
        
        # Break into time and voltage lists.
        t, v = df[:, 0], df[:, 1]
        x = np.linspace(start, stop, 1000)
        
        # Plot the fitted normal curve, and histogram.
        axis.plot(x, norm(x, v.mean(), v.std()), color='k', linewidth=1.5, label=f'{v.mean():.3f}($\pm${v.std():.3f})V')
        axis.hist(v, bins=bins, normed=True, edgecolor='k', color='lightyellow', label=label)
        
        # Set plot parameters.
        axis.legend(loc='upper right')
        axis.grid(linestyle=':', zorder=0)
    
    # Set global plot parameters.
    f.text(0.5, 0.04, 'Voltage (V)', ha='center')
    f.text(0.02, 0.5, 'Frequency, $f$', va='center', rotation='vertical')

    # Save if filename given
    if save:
        plt.savefig(save, dpi=300)
    
    # Show plot.
    plt.show()
```

```python
t = np.linspace(0, 1000, int(1e6))
v = np.sin(t)

x = np.linspace(-2, 2, 1000)

plt.hist(v, edgecolor='k', color='lightyellow', bins=40, normed=True, label='pure sine')
plt.plot(x, norm(x, v.mean(), v.std()), color='k', linewidth=1.5)
plt.grid(linestyle=':')
plt.xlabel('Voltage (V)')
plt.ylabel('Frequency, $f$')
plt.legend(loc='upper right')
plt.savefig('Pure Sine', dpi=300)
plt.show()
```

```python
from scipy import signal

file = '3 m 400 mV.csv'

df = pd.read_csv(file, header=None)
df = np.asarray(df)
v = df[:,1]*1000

i = np.random.normal(loc=v.mean(), 
                     scale=v.std(), 
                     size=len(v) * 10)

i = signal.resample(i, num=len(v))

v.sort()
i.sort()

m, b = np.polyfit(i, v, 1)
x = np.linspace(i.min(), i.max(), 1000)

plt.scatter(i, v, color='k', marker='o', s=1)
plt.plot(x, m * x + b, color='k', linestyle='-', linewidth=1)

plt.xlabel('Theoretical Quantiles (mV)')
plt.ylabel('Sample Quantiles (mV)')

plt.savefig('qqplot.png', dpi=300)
plt.show()

v
```
