First we need to import the necessary libraries. `pandas` make importing `*.csv` data easy, `numpy` is a collection of numerical methods written in fast FORTRAN, and `pyplot` helps with plotting.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

Next, we need to import our `*.csv` data into an array.

```python
data = pd.read_csv('test.csv', header=None)
data = np.asarray(data)
```

We can us `numpy` to quickly gather some single variable statistics on our data.

```python
mu    = data.mean()
sigma = data.std()

print("{mu:.1f} (±{sigma:.1f})".format(mu=mu, sigma=sigma))
```

Now we just need to plot. This makes Excel feel *very* painful!

```python
plt.hist(data, density=True, color='darkgrey')

x = np.linspace(65, 90, 1000)
z = (x - mu) / sigma

plt.plot(x, 1.0 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-z**2 / 2), linewidth=2, color='y')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('The Frequency Distribution of Normal Data')

plt.savefig('histogram.png')
```

See that last bit, the `plt.savefig()`. I just saved my graph as an image, and If I were working with a LaTeX document, that updated image would have been automatically updated in my report!
